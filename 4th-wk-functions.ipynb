## weight control 함수

```r
weight_control <- function(curr_weight, target_weight, name, maxcnt) {
  for( cnt in  1:maxcnt) {
    if (target_weight  >= curr_weight) break
    print("1,000 times jump ropes")
    curr_weight = curr_weight - 5.0 * runif(1)
  }
  return(curr_weight)
}
```

## 디폴트 인자값 

```r
weight_control <- function(curr_weight, target_weight, name, maxcnt=3) {
  for( cnt in  1:maxcnt) {
    if (target_weight  >= curr_weight) break
    print("1,000 times jump ropes")
    curr_weight = curr_weight - 5.0 * runif(1)
  }
  return(curr_weight)
}


names = c('Bill','Kim',"Park",'Son')
weights = c(80,90,100,75)
targets = c(70,75,80, 60)

print(weights)
for(i in 1:length(names)) {
  weights[i] = weight_control(weights[i],targets[i],names[i])
}  
cat("targets= ", targets,"\n")
cat("resulsting= ", weights)
```

## 함수내에서 전역 변수 변경하기 : <<-

## 함수 내에서 전역변수를 읽기 

```r
calc_area <- function (radius) {
  result <- 3.14 * r**2 + radius 
  return(result)
}

r <- as.numeric(readline(prompt='radius: '))
area <- calc_area(r)
print(area) 
```

## 전역 변수 변경?

```r
calc_area <- function (radius) {
  b <- 3.14 * r**2 * area + radius 
  area <- 3.14 * radius
  return(b)
}
area <- 0.0
r <- as.numeric(readline(prompt='radius: '))
result <- calc_area(r)
print(area)  
print(result)
```

```r
calc_area <- function (radius) {
  b <- 3.14 * r**2 * area + radius 
  area <<- 3.14 * radius
  return(b)
}
area <- 0.0
r <- as.numeric(readline(prompt='radius: '))
result <- calc_area(r)
print(area)  
print(result)
```

## 벡터에 원소 추가하기

```r
v = c()
w = c()
for(i in 1:10) {
  v = c(v, i)
  w = c(w, exp(i))
}
print(v)
print(w)
```

## 그외 벡터에 적용 가능한 유용한 함수들...
-  max(x), min(x), sum(x), var(x), sd(x)

-  prod() :  prod(x)  : 벡터 x의 모든 요소의 곱

-  range() :  range(x) : 벡터 x의 범위 (min, max)

-  rev() :  rev(x) : 벡터 x의 순서를 뒤집은 벡터 

-  summary() : summary(x) : 벡터 x의 요약 통계량 

- unique() :  unique(x)  : 중복 값이 제거된 벡터

-  cor() :  cor(x, y) :  벡터 x와 y의 상관계수

- cumsum() :  cumsum(x) : cumulative sum 벡터 x의 누적합을 구함. (벡터 결과)
 
- cumprod() :  cumprod(x) : cumulative product 벡터 x의 누적합을 구함. (벡터 결과)

- seq(start, end, step) : sequence generation

```r
x = c(3,5,6,7,9, 6, 6, 10, 15, 20, 5)
print(prod(x))

print(range(x))

print(rev(x))

print(summary(x))

print(unique(x))

y = runif(length(x))

print( cor(x,y))

print(cumsum(x))

print(cumsum(y))

print(cumprod(x))

print(seq(1,100,4))
```
