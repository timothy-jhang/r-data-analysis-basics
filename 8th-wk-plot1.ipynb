## 데이터 처리 패키지 사용하지 않고 piechart 그려 보기
- alcohol group의 비율을 파이차트로 가시화하기

```r
cnt <- rep(0,length(levels(esoph$alcgp)))
name <- rep(" ",length(levels(esoph$alcgp)))

print(cnt)
print(levels(esoph$alcgp))
level <- levels(esoph$alcgp)

x=esoph[esoph$alcgp=="40-79",]
print(x)
print(dim(x)[1])

for(i in 1:length(level)) {
   name[i] <- level[i]
   cnt[i] <- dim(esoph[esoph$alcgp==level[i],])[1] 
}
print(cnt)
str(esoph)
names(cnt) <- name # levels(...)
pie(cnt)
```
