## matrix 만들기

```r
sc2 <- matrix( c(50,60,70,30,70,90), nrow=2, byrow=TRUE) # by default, column-major
print(sc2)
```

```r
rownames(sc2) <- c('kim', 'lee')
colnames(sc2) <- c('math', 'art', 'lang')
print(sc2)
sc2
```

```r
print(matrix(1:6, nrow=2)) # 기본은 열 먼저 채우기!!  행의 개수  정해주기 nrow=

print(matrix(1:6, ncol=3)) # 기본은 열 먼저 채우기!!   열의 개수 정해주기  ncol=

print(matrix(1:6, nrow=2, byrow=TRUE)) # 행 먼저 채우기 byrow=TRUE

print(matrix(2:3, nrow=2, ncol=3)) # 행과 열의 개수 설정, 개수가 모자라면 자동 반복
# recycle 1:3

```

##rbind(), cbind() : row로 붙이기, column으로 붙이기

```r
print(cbind(1:3, 2:4))   # column 별로, 붙이기  cbind(  ) 

print(rbind(1:3, 5:7))  # row 별로,   붙이기  rbind(  )
```

```r
m <- matrix(1:6, byrow = TRUE, nrow = 2)
print(m)

print(rbind(m, 7:9)) # m의 마지막행에  행별 7:9를 붙임

print(cbind(m, c(10, 11))) # m의 마지막 열에  열별로  붙임
```

## rownames(), colnames()

```r
m <- matrix(1:6, nrow = 2, byrow=T) 
print(m)
rownames(m) <- c('r1','r2') #행 이름  지정
colnames(m) <- c('c1','c2','c3') #열 이름 지정
print(m)
```

```r
m <- matrix(1:6, nrow = 2, byrow=T, 
                   dimnames=list(c('r1','r2'), c('c1','c2','c3') ) ) 
print(m) 
```

## subset 취하기

```r
m <- matrix(sample(1:15,12),nrow=3)
print(m)

print(m[1,3])

print(m[1:2,2:3])

print(m[1,])

print(m[,2])

print(m[12])

print(m[8])

print(m[1,1:2])
```

## subset with c()

```r
m <- matrix(sample(1:15,12),nrow=3)
print(m)

print(m)
m
print(m[c(1,2),c(2,3)]) #1,3행의  2,4열  

print(m[c(2,3),c(4,1,2)]) #2,3행의  4,1,2열 

```

```r
rownames(m)<- c('r1','r2','r3') 
colnames(m)<- c('c1','c2','c3','c4') 
print(m) 

print(m[2,3])

print(m['r1','c3']) #1행, 3열

print(m['r2',c('c2','c4')])#2행, 2,3열

print(m[c('r1','r3'),c('c2','c4')])#1,3행, 2,4열
```

## colsums(), rowsums()

```r
sc2 <- matrix( c(50,60,70,30,70,90), nrow=2, byrow=TRUE)
sc2

rownames(sc2) <- c('kim', 'lee')
colnames(sc2) <- c('math', 'art', 'lang')
sc2

rowSums(sc2)
rowSums(sc2)/dim(sc2)[2]  # 사람별 평균 점수

colSums(sc2)
colSums(sc2)/dim(sc2)[1]  # 과목별 평균 점수 

sum(sc2)/length(sc2)  #전체 평균
```
