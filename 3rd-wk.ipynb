## 단독 원소는 벡터
- 인쇄된 모양으로 보면, 원소 한개와 원소 두개 차이가 없음.
- length(v) : 원소의 갯수 리턴

```r
my_apples <- 5
is.vector(my_apples)
print(my_apples)

my_oranges <- 'six'
is.vector(my_oranges)
print(my_oranges)
print(length(my_oranges))

my_orange2 <- c('six', 'seven')
is.vector(my_orange2)
print(my_orange2) # 인쇄된 모양으로 보면, 원소 한개와 원소 두개 차이가 없음. 
print(length(my_orange2))
```

## 벡터의 연산, 단일 상수와의 연산
- 단일 상수는 벡터와 같은 크기로 확장되어서 계산됨.
- 반복문을 사용한 것과 같은 효과가 발생함. 
- 벡터의 연산자에는 for() 반복문이 숨어 있음. 

```r
earn <- c(5, 10, 3)
print(earn)
earn1 <- earn * 34
print(earn1)
earn1 <- earn * c(34,34,34)
print(earn1)

#살제 계산은 반복문을 사용한 것과 같은 효과가 남.
earn2 <- earn
for(i in 1:length(earn2)) { earn2[i] <- earn2[i] * 34}
print(earn2)
```

```r
earn3 <- earn/3
print(earn3)

earn3 <- earn/c(3,3,3)
print(earn3)

#반복문 사용시.
earn4 <- earn
for(i in 1:length(earn4)) { earn4[i] <- earn4[i] / 3}
print(earn4)
```

```r
earn5 = earn - 3
print(earn5)

earn6 = earn ^ 2 
print(earn6)
```

## 벡터 끼리의 연산에서도
- 많은 반복문을 연산자로 줄여서 사용할 수 있음. 

```r
# 많은 반복문을 연산자로 줄여서 사용할 수 있음. 
print( earn5 * earn6 / 2.5 + earn * earn2 - earn3 / earn4 * 3)
```

## 벡터용 함수, 연산자들
- sum(), max(), min(), mean(), var(), 표준편차, 
- 많은 반복문이 포함되어 있음.
- 뎃셈, 최대값찾기, 최소값찾기, 평균값 찾기, 분산계산, 표준편차계산
- v > 0.5 : 크기 비교 (각 원소들과 0.5와 크기 비교) 

```r
v = runif(100)
print(v)
sum(v)
max(v)
min(v)
mean(v)
var(v)
sd(v)
print(v > 0.5)
print(v[v > 0.5])
```

## logical을 이용한 서브셋 취하기 - subset by logical vectors


```r
remain <- c(40,20,30,70,90, 15)
names(remain) <- c('a','b','c','d','e','f')
print(remain)
print(remain > 30)
print(remain > c(30,30,30,30,30,30) )

print(remain[remain>30])
print(remain[remain<30])
print(remain[remain>30 & remain<90])

print(remain[c(T,F)]) # # recycle c(T,F), 벡터가 반복됨!
```
