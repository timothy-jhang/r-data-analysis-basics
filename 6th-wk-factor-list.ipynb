##factor
- factor()는 category 변수에 대해서, character를 factor (수자)로 변환함
- factor() 후에는 정수 값이 됨. (Factor level - 값)
- factor(, levels=…), 또는 levels(factor) = …. 형태로 지정가능 (값을 바꿈)
- Labels= 인자로 level 이름을 바꿀 수도 있음. (이름을 바꿈)
- 순서가 있는( >,<로 크기 비교 가능) factor를 만들려면,  factor(, ordered=TRUE) 를 사용함.


- 작성자: [충남대학교 인공지능학과 장경선](http://www.cnu.ac.kr/~sun)

```r
bloodtype <- c("B", "AB", "O", "A", "O","O","A","B","O","AB")
bloodtype

bloodfactor <- factor(bloodtype)
bloodfactor

str(bloodfactor)

str(bloodtype)
```

```r
bloodfactor2 <- factor(bloodtype, levels=c("O","A","B","AB"))
bloodfactor2

str(bloodfactor2)

str(bloodfactor)

```

```r
bloodtype

bloodfactor

levels(bloodtype) # not a factor

levels(bloodfactor)

levels(bloodfactor2)

print('-------- levels of bloodfactor ---')
levels(bloodfactor) <- c('RH+A','RH+AB', 'RH+B','RH+O')

bloodfactor

str(bloodfactor)

print('------bf3----')
bf3 <- factor(bloodtype, labels=c('RH+A','RH+AB', 'RH+B','RH+O'))
bf3

str(bf3)

print('-------- bf4 ----')
bf4 <- factor(bloodtype, labels=c('RH+A','RH+AB', 'RH+B','RH+O'),levels=c('O','A','B','AB'))
bloodtype
bf4
str(bf4)
```

## ordinal

```r
tz <- c('M','S','M','L','S','M','XL')
tz2 <- factor(tz, ordered=T, levels=c('S','M','L','XL')) 

str(tz2)

str(tz)

tz1 <- factor(tz)
str(tz1)

tz2[1] > tz2[2]

tz1[1] > tz1[2]   #ordered factor가 아님.
```

##[], [[]] - 차이?
- 리스트의 각 원소는 리스트 임. 
- [i] : 리스트의 i번째 원소 - 역시 리스트 임.
- [[i]] : 리스트의 i번째 원소의 원래 내용을 취함 : [] 리스트의 내용을 까서 그 안에 내용을 취한다는 의미로 생각!

```r
song1 <- list('peace', 1999, 5)
song1[1]
print('-----------')
str(song1[1])
print('-----------')
str(song1[[1]])
```

```r
song1 <- list('peace', 1999, 5)
names(song1) <- c('title','year', 'trks')

cd1 <- list(title='peace and blue', year=1999, trks=10, song=song1)
cd1
cd1['song']          #해당 열 자체 - 리스트
print('------------')
cd1[['song']]        #해당 열의 내용

print('------------')
cd1[4]
print('------------')
cd1[[4]]
```

```r
length(cd1)
str(cd1)

print("cd1[[4]][2]==>") 
print(cd1[[4]][2]) 
print("cd1[[4]][[1]]==>")
print(cd1[[4]][[1]])

print('-----')
print(cd1[[4]][3])
print('---')
print(cd1[3])
print(cd1[[3]])
```

## 리스트 실습하기
- 가.  mon, tue, wed, thu, fri, sat, sun 로 구성된 day
- 나. 1, 2, 3, …, 31 로 구성된 date
- 다. Jan, feb, mar, apr, may, jun, jul, aug, sep, oct, nov, dec 로 구성된  month
- 라. 1:3000으로 구성된 year
1. 위 네가지 vector로 구성된  list 변수  L1 만들기
2. day, month는  factor화 하기. + ordered factor + levels가 우리가 원하는 순서대로 ...
3. 2022년 4월 12일 화요일 정보를  L1에 저장된 정보를 기반으로 표시하려면?  


```r
day = c('mon','tue', 'wed', 'thu', 'fri', 'sat', 'sun')
day = factor(day, ord=T, levels=c('mon','tue', 'wed', 'thu', 'fri', 'sat', 'sun'))
date = c(1:31)

month=c('jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec')
month = factor(month, ord=T,levels=c('jan','feb','mar','apr','may','jun','jul','aug','sep','oct','nov','dec'))

year = c(1:3000)
L1 = list()
L1$day = day
L1$date = date
L1$month = month
L1$year = year

L1[[1]][2] 
L1[[2]][12]
L1[[3]][4]
L1[[4]][2020]
```
