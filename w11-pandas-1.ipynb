###pandas series, pd.Series(list) 
- 열에 해당함. 
- 열의 각 행의 접근은 index 정의 전에는, 0, 1, 2와 같은 정수로 접근
- index 정의 후에는 index label에 따라 접근가능함.

```python
import pandas as pd

a = [1, 7, 2, 10, 13, 12]

myvar0 = pd.Series(a) # before setting index

print(f'myvar0=\n{myvar0}')
print(f'myvar0[2]={myvar0[2]}')

myvar = pd.Series(a, index = ["t", "u", "w", "x", "y", "z"])

print(f'myvar=\n{myvar}')
print(f'myvar["w"]=\n{myvar["w"]}')
print(f'myvar[3]=\n{myvar[3]}')
```

###pandas dataframe, pd.DataFrame( data )

```python
# list -> dataframe
import pandas as pd

data1 = [1,2,3,4]
df1 = pd.DataFrame(data1, columns=['v1'],index=['a','b','c','d'])
print(f'df1=\n{df1}')
print(f'df1.index={df1.index}')
print(f'df1.loc[a]=\n{df1.loc["a"]}')
print(f'df1.loc[a][0]=\n{df1.loc["a"][0]}')

print(f'df1["v1"]=\n{df1["v1"]}')
print(f"df1['v1'].loc['a']=\n{df1['v1'].loc['a']}")
```

## dictionary에서 dataframe 생성하기
- dictionary의 key는 column label로 사용됨.
- row label의 기본값은 0,1,2,3, ...
- row label 지정은  index 인자로 가능함. 


```python
# dictionary to dataframe
import pandas as pd
data = {
  "calories": [420, 380, 390],
  "duration": [50, 40, 45]
}
df0 = pd.DataFrame(data) # dictionary has column labels

print(f'df0=\n{df0}')

print(f'df0.loc[0]=\n{df0.loc[0]}')
print(f'df0.loc[0][0]=\n{df0.loc[0][0]}')
print(f'df0.loc[0][1]=\n{df0.loc[0][1]}')
print(f"df0.loc[0]['calories']=\n{df0.loc[0]['calories']}")

df = pd.DataFrame(data, index = ["day1", "day2", "day3"])

print(f'df=\n{df}')

print(f'df.loc["day1"]=\n{df.loc["day1"]}')
# index가 만들어진 후에는, 기본값 0,1,2를 사용하면 오류 발생됨.
#print(f'df.loc[1]=\n{df.loc[1]}')

print(df['calories'])
print(df['duration'])
print(type(df['duration']))
print(df['duration']['day2'])
print(df.loc['day2','duration'])
```


## df.append() : insert row to df
- DataFrame.append(other, ignore_index=False, verify_integrity=False, sort=False)[source]¶
- other 인자에 있는 행을 dataframe에 추가, 새로운 오브젝트 리턴.
- other : DataFrame 또는 Series나 dict-like 객체 : 추가할 데이터들을 가리킴
- ignore_indexbool : default False, 참이면, 0,1,2,...,n-1로 레이블링됨.

```python
import pandas as pd
df = pd.DataFrame([[1, 2], [3, 4]], columns=list('AB'), index=['x', 'y'])
print(f"df=>\n{df}")
df2 = pd.DataFrame([[5, 6], [7, 8]], columns=list('AB'), index=['x', 'y'])

print(f"df2=>\n{df2}")
print(f"df.append(df2)=>\n{df.append(df2)}")

df3 = df.append(df2, ignore_index=True)
print("df3 = df.append(df2, ignore_index=True)")
print(f"df3=>\n{df3}")

# less efficient
df4 = pd.DataFrame(columns=['A'])
print(f"df4=>\n{df4}")

for i in range(5):
    df4 = df4.append({'A': i}, ignore_index=True)
    print("after appending ", i)
    print(f"df4=>\n{df4}")
print(f"df4=>\n{df4}")

# more efficient
df5 = pd.concat([pd.DataFrame([i], columns=['A']) for i in range(5)],
          ignore_index=True)
print(f"df5=>\n{df5}")

# dict-like 
df6 = df5.append( other={'A':7 }, ignore_index=True)
df6.head(8)
```

### df.insert(pos, ...)

- loc : int, 삽입 인덱스 위치, 값의 범위는 0 <= loc <= len(columns).
- column : str, number, or hashable object, 삽입 열의 이름

## pd.concat(objs, axis=0, ...)
- pd.concat(objs, axis=0, join='outer', ignore_index=False, keys=None, levels=None, names=None, verify_integrity=False, sort=False, copy=True)
- pd object를 concat 수행 - 정해진 축으로 (기본축은 행 방향으로 접합)
- 접합 축에 따라 계층적 인덱스 추가 가능 : keys 인자 사용
- ignore_index - 기존 index 초기화
- names - 열 이름


```python
# pd.concat( ( ), axis)
import pandas as pd
s1 = pd.Series(['a', 'b'])
s2 = pd.Series(['c', 'd'])
print(f"s1=>\n{s1}")
print(f"s2=>\n{s2}")
print(f"pd.concat([s1, s2])=>\n{pd.concat([s1, s2])}")
print(f"pd.concat([s1, s2], axis=1)=>\n{pd.concat([s1, s2], axis=1)}")
print(f"pd.concat([s1, s2], ignore_index=True)=>\n{pd.concat([s1, s2], ignore_index=True)}")
print(f"pd.concat([s1, s2], keys=['s1','s2'] )=>\n{pd.concat([s1, s2], keys=['s1','s2'] )}")
s1s2 = pd.concat([s1, s2], keys=['s1','s2'])

print("s1s2 = pd.concat([s1, s2], keys=['s1','s2'])")
print(f"s1s2['s1'][0]=>\n{s1s2['s1'][0]}")
print(f"s1s2['s1'][1]=>\n{s1s2['s1'][1]}")

print(f"pd.concat([s1, s2], keys=['s1','s2'], axis=1)=>\n{pd.concat([s1, s2], keys=['s1','s2'], axis=1)}" )

print(f"pd.concat([s1, s2], keys=['s1','s2'], names=['Series name', 'Row ID'] )=>\n{pd.concat([s1, s2], keys=['s1','s2'], names=['Series name', 'Row ID'] )}" )
```

```python
df1 = pd.DataFrame({'col1': [1, 2], 'col2': [3, 4]})
print(f'df1=>\n{df1}')
df2 = pd.DataFrame({'col1': [5, 6], 'col2': [7, 8]})
print(f'df2=>\n{df2}')
print(f"pd.concat([df1,df2],ignore_index=True) => \n{pd.concat([df1,df2],ignore_index=True) }")
print(f"pd.concat([df1,df2],axis=1) => \n{pd.concat([df1,df2],axis=1) }")
print(f"pd.concat([df1,df2]) => \n{pd.concat([df1,df2]) }")
```

### df.cummax(), df.cummin(), df.cumsum(), df.cumprod()
- series에 대해서, dataframe의 row에 대해서, 적용
- s.idxmax(), s.idxmin() : index of max/min


```python
s = pd.Series([2, 3, 5, -1, 0])
print(f"s=>\n{s}")

print(f"s.cummax()=>\n{s.cummax()}")
print(f"s.cumsum()=>\n{s.cumsum()}")
print('---\n',s.max(), s.min(), s.sum(), s.mean())
print('---\n',s.idxmax(), s.idxmin())
```

```python
s = pd.Series([2, np.nan, 5, -1, 0])

print(f"s=>\n{s}")

print(f"s.cummax()=>\n{s.cummax()}")
print(f"s.cummax(skipna=False)=>\n{s.cummax(skipna=False)}")

print('---\n',s.max(), s.min(), s.sum(), s.mean())
print('---\n',s.idxmax(), s.idxmin())
```

```python
df = pd.DataFrame([[2.0, 1.0],
                   [3.0, np.nan],
                   [1.0, 0.0]],
                   columns=list('AB'))

print(f"df=>\n{df}")

print(df.cummax())
print(df.cummax(axis=1))

print('----\n',df.min())
print('----\n',df.max())
print('----\n',df.idxmin())
print('----\n',df.idxmax())

print('----\n', df.sum(), df.mean())
```

### df.items() : 열 별로 데이터 살펴보기 가능 

```python
df = pd.DataFrame({'species': ['bear', 'bear', 'marsupial'],
                  'population': [1864, 22000, 80000]},
                  index=['panda', 'polar', 'koala'])
for label, content in df.items(): # 열별로 데이터와 컨텐츠 처리 / dict 처리와 같음.
    print(f'label==> \n{label}')
    print(f'content==> \n{content}', sep='\n')
    print('-------------------------------')
```

```python
df = pd.DataFrame(np.arange(10).reshape(-1, 2), columns=['A', 'B'])
print('-----\n',f'df=>\n{df}')
m = df % 3 == 0
print("m = df % 3 == 0")
print('-----\n',f"df.where(m,-df)=>\n{df.where(m,-df)}")

print('-----\n',f"df.where(m, -df) == np.where(m, df, -df) =>\n{df.where(m, -df) == np.where(m, df, -df)}" )
print('-----\n',f"df.where(m, -df) == df.mask(~m, -df) =>\n{df.where(m, -df) == df.mask(~m, -df)}" )
```

### pandas.DataFrame.corr(method='pearson')
- column 들간의 correlation 계산,  NA/null값은 제외
-  method{‘pearson’, ‘kendall’, ‘spearman’}
- > pearson : standard correlation coefficient
- > kendall : Kendall Tau correlation coefficient
- > spearman : Spearman rank correlation

```python
df = pd.DataFrame([(.2, .3, .4), (.0, .6, 0.7), (.6, .0, 0.9), (.2, .1, .3)],
                  columns=['dogs', 'cats', 'men'])
print('-----\n',f'df=>\n{df}')
print('-----\n',f"df.corr(method='pearson')=>\n{df.corr(method='pearson')}")
print('----\n',f"df.corr(method='kendall')=>\n{df.corr(method='kendall')}")
print('----\n',f"df.corr().loc['dogs','cats']=>\n{df.corr().loc['dogs','cats']}")
print('----\n',f"df.corr(method='kendall').loc['dogs','cats']=>\n{df.corr(method='kendall').loc['dogs','cats']}")
```

### pd.DataFrame.apply(func, axis=0 )
- funcfunction : 각 column/row에 적용할 함수
-  axis{0 or ‘index’, 1 or ‘columns’}, default 0, 
함수가 적용되는 축
- > 0 or ‘index’: 각 열에 대해 적용 (각 열의 모든 행에 대해서 적용)
- > 1 or ‘columns’: 각 행에 대해 적용 

```python
import pandas as pd
import numpy as np
df = pd.DataFrame([(.2, .3, .4), (.0, .6, 0.7), (.6, .0, 0.9), (.2, .1, .3)],
                  columns=['dogs', 'cats', 'men'])
df.corr()
print(f'df=>{df}')
print('---\n',f"df.apply(np.sum, axis=0)=>\n{df.apply(np.sum, axis=0)}" )
print('---\n',f"df.apply(np.mean, axis=1)=>\n{df.apply(np.mean, axis=1)}" )
print('---\n',f"df.apply(np.mean, axis=1)[0:2]=>\n{df.apply(np.mean, axis=1)[0:2]}")
```
