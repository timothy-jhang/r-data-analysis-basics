- Long form – 저장하기 간편하고, 변환이 쉬움.
-- 한 열에 여러 measure(측정)치를 가질 수 있음.
-- 그래서, 열이 길어짐.
- Wide form – 원래 data.frame 같은 형태, 비교가 쉬움, 그래프 그리기 편함.
-- 한 열은 하나의  measure(측정) 변수
-- 한 행은 하나의 관찰(observation), 여러 measure(측정)치를 가짐.

# Rectangling
- 깊게 nesting된 리스트(JSON이나 XML에서 가져온 경우가 많음)를 행들, 열들로 tidy하게 만듬.
- unnest_longer() : list-column의 각 리스트 원소를  행으로 만듬
- unnest_wider() : list-column의 각 리스트 원소를 열로 만듬.
- unnest_auto() : unnest_longer()와 unnest_wider() 중 적절한 동작 수행
- hoist() : unnest_wider()와 유사하나, 선택된 요소만 unnest 함. 

## hoist(), unnest_longer(), unnest_wider()

```r
library(tidyverse)
df <- tibble(
  character = c("Toothless", "Dory"),
  metadata = list(
    list(
      species = "dragon",
      color = "black",
      films = c(
        "How to Train Your Dragon",
        "How to Train Your Dragon 2",
        "How to Train Your Dragon: The Hidden World"
       )
    ),
    list(
      species = "blue tang",
      color = "blue",
      films = c("Finding Nemo", "Finding Dory")
    )
  )
)
str(df)
df
```

```r
df %>% unnest_wider(metadata)
```

```r
df %>% unnest_longer(metadata)
```

```r
str(df)
df %>% unnest_auto(metadata)
```

```r
# Extract only specified components
df1 <- df %>% hoist(metadata,
  "species",
  first_film = list("films", 1L),
  third_film = list("films", 3L)
)
str(df)
df1
str(df1)
```

```r
str(df) # 두개의 리스트 요소 가짐. 

df3 <- df %>% unnest_wider(metadata)
df3
str(df3) # 한개의 리스트 원소 가짐. 

# 남은 리스트 원소를 풀어 냄. 일반적인 pivot_wider()와 유사하지만, 구분됨.
df4 <- df %>% unnest_wider(metadata) %>% unnest_longer(films)
df4
str(df4)
```

```r
# nest 부터, 시작해서, unnest_wider 까지 해보기
df <- tibble(x = c(1, 1, 1, 2, 2, 3), y = 1:6, z = 6:1)
str(df)

df5 <- df %>% nest(cols=c(x,y)) 
str(df5)
df5

df6 <- df %>% nest(cols=c(x,y)) %>% unnest_wider(cols)
str(df6)
df6
```

## [separate()]( https://tidyr.tidyverse.org/reference/separate.html)
- 문자로된 열을 여러 열로 나누기
- separate(data,col,into,sep = "[^[:alnum:]]+",remove = TRUE,convert = FALSE, extra = "warn",fill = "warn",...)
-- col : 분리 대상이 되는 열
-- into : 분리 후 만들어질 열들의 이름 (문자 벡터)
-- remove 인자 : 기본값 TRUE : 입력 col은 출력 dataframe에서는 삭제됨.
-- convert 인자 :  TRUE인경우 integer, numeric, logical은 변환까지 수행.
-- extra 인자 : sep이 문자벡터이고, 열 수보다 많은 경우에, 3가지 옵션값이 있음 :  "warn" (the default): 경고 내보내고 남는 값은 drop, "drop"은 경고 없이 남는 값은 drop, "merge"은 남는 값들은 병합해서 하나로 만듬. 
-- fill 인자 : sep이 문자 벡터이고, 열 수보다 적은 경우에 3가지 옵셥값이 있음 : "warn" (the default): 경고보내고 모자란 값은 NA로 채우며, 오른쪽 부터 채움, "right": 경고 없이 오른쪽부터 NA로 채움. "left": 경고 없이, 왼쪽부터 NA로 채움.

```r
dftime <- data.frame(time=c("12:30", "10:20", "11:36","13.17"))
dftime
dftime %>% separate(col=time,into=c("hour", "minute"), convert=T) # sep이 없어서 알아서...
```

```r
df <- data.frame(x = c(NA, "x.y", "x.z", "y.z"))
df %>% separate(x, c("A", "B"))
```

```r
# If you just want the second variable: 둘째 변수/열 B만 필요한 경우.
df %>% separate(x, c(NA, "B"))

dftime %>% separate(time,c(NA,"minute"))
```

```r
df <- data.frame(x = c("x", "x y", "x y z", NA))
df %>% separate(x, c("a", "b")). 
```

```r
df %>% separate(x, c("a", "b"), extra = "drop", fill = "right") # 경고 없음. 결과는 동일 
```

```r
df %>% separate(x, c("a", "b"), extra = "merge", fill = "left") # fill은 왼쪽부터, extra는 병합
```

```r
df %>% separate(x, c("a", "b", "c"),fill='right')
```

```r
df <- data.frame(x = c("x: 123", "y: error: 7"))
df %>% separate(x, c("key", "value"), ": ", extra = "merge") # sep 사용됨. 
```

```r
df <- data.frame(x = c(NA, "a1b", "c4d", "e9g"))
df %>% separate(x, c("A","B"), sep = "[0-9]")  # sep로 숫자 사용
```

```r
df <- data.frame(x = c("x:1", "x:2", "y:4", "z", NA))
df %>% separate(x, c("key","value"), ":") #%>% str       # sep로 colon 사용
```

```r
df %>% separate(x, c("key","value"), ":", convert = TRUE) #%>% str  # 숫자는 변환까지, convert=TRUE
```

## complete() -  데이터프레임/테이블에서 모든조합을 포함하도록 만들어감 
- https://tidyr.tidyverse.org/reference/complete.html
- 빠진 조합을 완성하는데 유용
- complete(data, ..., fill = list(), explicit = TRUE)
- ... : 모든 조합을 생성할 열들
- fill : named list (각 변수에 대해 NA대신 사용할 값 하나 씩 지정)
- explicit = TRUE/FALSE : fill 동작시에, 원래부터 있던 NA도 fill 값으로 채움/안채음.



```r
df <- tibble(
  group = c(1:2, 1, 2),
  item_id = c(1:2, 2, 3),
  item_name = c("a", "a", "b", "b"),
  value1 = c(1, NA, 3, 4),
  value2 = 4:7
)
df

df %>% complete(group, item_id, item_name) # 3개 열의 모든 가능한 조합을 만들고 빠진 조합의 해당 열 값은  NA 채우기 
```

```r
# NA 대신 사용할 값 채우기, fill = list( col1=val1, col2=val2)
df %>% complete(group, item_id, item_name, fill = list(value1 = 0, value2 = 99))
```

```r
# 모든조합이 아닌, item id, item_name은 이미 있는 조합을 묶어서 패턴으로 사용
df %>% complete(group, nesting(item_id, item_name))
```

```r
# group과 item_id만 활용한 조합 만들기 
df %>% complete(group, item_id)
```
