## %%R : 현재 cell을 R로 코딩할 수 있게 해줌.

```python
%%R
v <- c(1:100)
print(v)
print(v[v > 50])

x <- seq(0, 2*pi, length.out=50)
print("x==>>")
print(x)
m1 <- matrix(x, nrow=5)
print("m1==>>")
print(m1)
m2 <- matrix(x, nrow=10)
print("m2==>>")
print(m2)
plot(x,sin(x))
hist(sin(x),10)
```

##Single magic R
- %R : 한문장만 R 코드 삽입 가능하게 해줌. 

```python
x = %R x
print(x[:20])
print(x[20:40])
m2 = %R m2
print(f'm2=>\n{m2}')
```
