## 요약하기 : summarise(), summarize()
- 그룹화(group_by)된 경우에, 그룹별 summary를 몇개의 행으로 생성하거나, 그룹화 안된 경우에는 단일 행으로 summary 생성 - 생성 결과는 새 데이터 프레임으로 만듬.
- 생성된 데이터 프레임의 열은 각 그룹 변수 이름과 그에 대한 summary 통계값 열을 포함함.

- **유용한 함수들**
- - Center: mean(), median()
- - Spread: sd(), IQR(), mad()
- - Range: min(), max(), quantile()
- - Position: first(), last(), nth(),
- - Count: n(), n_distinct()
- - Logical: any(), all()

```r
library(tidyverse)
# 그룹화 안된 경우 - 한 줄 요약
mtcars %>% summarise(mean = mean(disp), n = n()) 
```

```r
# group_by() 로 그룹화 한 경우, 그룹별 요약
mtcars %>%
  group_by(cyl) %>%
  summarise(mean = mean(disp), n = n())
```

```r
mtcars %>%
  group_by(cyl, vs) %>%
  summarise(cyl_n = n())
```

# 모든 열, 조건에 맞는 열 , 선택 열에 대한 요약
-  summarise_all, summarise_if, summarise_at

### summarise_all() 모든 열에 대해서 적용

```r
# Diet로 그루핑, Chick열과 Time열 삭제, 나머지(weight)열에 대해서, min, max, mean 열 추가
ChickWeight %>% group_by(Diet) %>% select(c(-Chick,-Time)) %>% summarise_all(list(min=min,max=max,mean=mean))

# remove rows with NA first, ... 
starwars %>% na.omit() %>% group_by(sex) %>% select( c(height, mass)) %>% summarise_all(list(min=min,max=max,mean=mean))
```

### summary_at(변수들,요약함수)

```r
starwars %>% na.omit() %>% group_by(skin_color,eye_color) %>% summarise_at(vars(height,mass), list(max1=max, min1=min, mean1=mean)) #%>% slice(1:5)
```

### summary_if(조건함수, 요약함수) 
- 조건함수 : 적용 열 선택 함수
- 요약함수 : 적용 열에 적용하는 함수

```r
starwars %>% na.omit() %>% group_by(skin_color,eye_color) %>% summarise_if(is.numeric, mean) 
```

# 모든 열, 조건에 맞는 열 , 선택 열에 대한 열변형/열추가하기
- mutate_all, mutate_if, mutate_at
- [참고 페이지](http://rstudio-pubs-static.s3.amazonaws.com/491966_5072cd8affd84069833a65aed85303b8.html)
- mutate_all : 해당 열에 모두 적용
- mutate_if : 특정 조건을 만족하는 열에 (예 is.numeric ) 적용
- mutate_at : 특정 선택 열에 적용 

```r
library(tidyverse)
```

### dataset to use

```r
water.quality.df <- read.csv(url("https://raw.githubusercontent.com/rg563/SPRING2019TIDYVERSE/master/beach-water-quality-automated-sensors.csv"),header=TRUE)
water.quality.tbl <- as_tibble(water.quality.df)
head(water.quality.tbl) #%>% select(1:5)
str(water.quality.tbl)
```

### mutate_all() 모든 열에 적용하기 

```r
# mutate_all
divide.by.100 <- function(x, na.rm=FALSE) (x/100)
water.quality.tbl.div.100 <- mutate_all(water.quality.tbl[,3:8], divide.by.100)
head(water.quality.tbl.div.100)

water.quality.tbl[,3:8] %>% mutate_all(divide.by.100) %>% slice(1:5)
water.quality.tbl %>% select(3:8) %>% mutate_all(divide.by.100) %>% slice(1:5)
```

### 조건에 맞는 열에만 적용하기 
- mutate_if(cond, if-function, else-function,...)

```r
# double 유형인 열에만 적용함. 
water.quality.tbl %>% slice(1:5)
water.quality.tbl %>% mutate_if(is.double, divide.by.100, na.rm = TRUE) %>% slice(1:5) #%>% select(2:5)
```

### 선택된 열에만 mutate 함수 적용하기 - mutate_at()

```r
#특정 선택 열에만 적용
convert.to.farenheit <- function(x, na.rm=FALSE) (x*9/5+32)
water.quality.tbl %>% mutate_at(c("Water.Temperature"), convert.to.farenheit) %>% slice(1:5) %>% select(1:4)
```
