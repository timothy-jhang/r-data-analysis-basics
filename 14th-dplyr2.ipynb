## KEY
- 테이블의 각 pair를 연결하는데 사용되는 변수/열  값을 key라고 함.
- Key는 한 변수일 수도 있고, 여러 변수 일수도 있다. 해당 observation을 유일하게 구분하는 것.
- 각 plane은 tailnum으로 유일하게 식별 가능
- Weather 테이블에서 각 observation은 year, month, day, hour, origin 모두를 사용해야 key가 될 수있다. 
- primary key : 소속 테이블의 각 observation을 유일하게 식별하는 key,  예: planes의 tailnum
- A foreign key : 다른 테이블의 각 observation을 유일하게 식별하는 key
- flights의 tailnum는 planes table의  행을 유일하게 식별하는 flights의 열로써 foreign key가 됨. 

- 한 변수가 primary key이면서, foreign key가 될 수도 있다.
- Origin 변수는 weather에서는 primary key의 일부이면서, Airports 에 대해서는 foreign key가 됨.

## mutating joins
- inner_join(x,y) : key가 양쪽 모두에 있는 observation만 포함
- left_join(x,y) : x에 있는 key의 모든 observation은 포함. 
- right_join(x,y) : y에 있는 key의 모든 observation은 포함. 
- full_join(x,y) : x와 y에 있는 모든 key의 모든 observation이 포함. 


```r
library(tidyverse)
x <- tribble(
  ~key, ~val_x,
     1, "x1",
     2, "x2",
     3, "x3"
)
y <- tribble(
  ~key, ~val_y,
     1, "y1",
     2, "y2",
     4, "y3"
)

x %>% 
  inner_join(y, by = "key")
```

## left_join, right_join, full_join 
- figure source : :https://r4ds.had.co.nz/relational-data.html
- ![left_join, right_join, full_join](https://drive.google.com/uc?id=1HZs7kMZAV2DHwb0Z3iHZMSf9ZEWy7hQZ)

## left_join(x,y)


```r
x <- tribble(
  ~key, ~val_x,
     1, "x1",
     2, "x2",
     3, "x3"
)
y <- tribble(
  ~key, ~val_y,
     1, "y1",
     2, "y2",
     4, "y3"
)

left_join(x, y, by = "key")
```

## right_join(x,y)

```r
x <- tribble(
  ~key, ~val_x,
     1, "x1",
     2, "x2",
     3, "x3"
)
y <- tribble(
  ~key, ~val_y,
     1, "y1",
     2, "y2",
     4, "y3"
)

right_join(x, y, by = "key")
```

##full_join(x,y)

```r
x <- tribble(
  ~key, ~val_x,
     1, "x1",
     2, "x2",
     3, "x3"
)
y <- tribble(
  ~key, ~val_y,
     1, "y1",
     2, "y2",
     4, "y3"
)

full_join(x, y, by = "key")
```

## 중복되어 나타나는 key의 처리
- 중복되는 나타내는 x의 observation에 대해, y의 대응 observation을 모두 포함시킴.

```r
# x에 나타나는 중복 key
x <- tribble(
  ~key, ~val_x,
     1, "x1",
     2, "x2",
     2, "x3",
     1, "x4"
)
y <- tribble(
  ~key, ~val_y,
     1, "y1",
     2, "y2"
)
left_join(x, y, by = "key")
```

```r
# x, y에 모두 나타나는 중복 key
x <- tribble(
  ~key, ~val_x,
     1, "x1",
     2, "x2",
     2, "x3",
     3, "x4"
)
y <- tribble(
  ~key, ~val_y,
     1, "y1",
     2, "y2",
     2, "y3",
     3, "y4"
)
left_join(x, y, by = "key")
```

## [filtering join : (X, Y)](https://rdrr.io/cran/dplyr/man/filter-joins.html)
- X의 observation에서 filter 동작을 하는데, Y와 key값의 매칭 여부에 따라 X의 observation을 선택함. 
- semi_join(), anti_join()
- semi_join() : Y 테이블과 매칭되는 Observation만 남김.
- anti_join() : Y테이블과 매칭안되는 Observation만 남김.

## [set operation](https://dplyr.tidyverse.org/reference/setops.html)
- https://dplyr.tidyverse.org/reference/setops.html
- intersect()
- union()
- setdiff()

```r
library(tidyverse)
df1 <- tibble(x = c(1,1,3,4), y=1:4)
df2 <- tibble(x= c(1,3,5,7,9), y=c(2,3,4,5,7))
df1
df2
print('intersect===>')
intersect(df1,df2)
union(df1,df2)
setdiff(df1,df2)
```
