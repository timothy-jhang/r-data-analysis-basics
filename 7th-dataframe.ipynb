## DataFrame
- 여러 개의 벡터로 구성됨. 
- 각 벡터는 하나의 열을 차지함.
- 2차원 데이터(Matrix)로 생각할 수 있음. 
- 하나의 열을 접근하기 :   aq[ ,1] ,     aq[, ‘Ozone’],열은 하나의 측정기준, 변수라고도 함.
- 하나의 행을 접근 하기 :  aq[1, ] :  하나의 관측/측정 (몇월몇일의 측정 데이터)

```r
aq = airquality
dim(aq)
str(aq)
```

##DataFrame의 열을 접근하는 방법들

```r
# 한 열을 인덱스로 (2차원 매트릭스에서 처럼)
aq[,1]
str(aq[,1]) #vector
```

```r
# 한 열을 이름으로 (2차원 매트릭스에서 처럼)
aq[,'Ozone']
str(aq[,'Ozone'])
```

```r
# 한 열을 $와 이름으로 접근 (리스트에서 처럼)
aq$Ozone
str(aq$Ozone)
```

```r
# 한열을 [[]]와 이름으로 접근 (리스트에서 처럼)
aq[['Ozone']]
str(aq[['Ozone']])
```

```r
# 한열을 DataFrame 그 자체로 접근하기 (리스트에서 처럼)
aq['Ozone']
str(aq['Ozone']) #dataframe
```

## 벡터들로 데이터프레임구성하기 

```r
ID <- 1:20
SEX <- sample( c("M","F"), size=20, replace=TRUE)
AGE <- sample( 20:60, size=20 )
AREA <- sample(c("서울","경기","제주","인천","대전","세종"), size=20, replace=T)
df <- data.frame(ID,SEX,AGE,AREA)
str(df)
dim(df)
## 자동 factoring이 되지는 않음. 
```

```r
## factor()이용 category변수를 문자에서 숫자로
df$SEX = factor(df$SEX)
df$AREA = factor(df$AREA)
str(df)
```

```r
## 자동factoring을 피해야 할 경우에... stringAsFactors=F
dataframe2 <- data.frame(id=ID, sex=SEX, age=AGE, area=AREA, stringsAsFactors = F)
str(dataframe2)
```

## 열 이름 붙이기

```r
str(df)
names(df) <- c('id','sex','age','area')
str(df)
```

## 열, 행에 접근 하기

```r
# 1-5행 접근
df[1:5,] # dataframe
```

```r
# 1-5행, 2열 접근
df[1:5,2]  # 벡터
```

```r
# 1-3행, 열을 이름으로 접근
df[1:3, 'age']
```

```r
# 원소 접근 시에는  [[]]도 사용도 무방
df[[1, 'age']]
```

##subset 취하기

```r
# 1-3행, 여러 열을 이름으로 접근
df[1:3, c('age','sex','area')] # dataframe
```

```r
## 행 일부, 열 일부 - subset
df[c(1,2,4),c(1,2,3)]
```

## logical을 사용한 subset 취하기

```r
#비교 연산 결과를 logical로 저장하기
cond = df$age > 50
cond
#logical의 결과의 길이는 df의 행 차원과 같음.
length(cond)
dim(df)
```

```r
## logical을 적용한 subsbet
df[cond, ]
```

```r
# 직접 logical 연산을 행 부분에 넣어서 처리하기.
df[df$age > 50,]
```

```r
# 좀더 복잡한 비교식도 가능함.
df[df$age > 50 & df$sex == 'F', ]
```

## 열 추가 하기 
- $ 이용 (리스트 처럼) 
- [[]] 이용 (리스트 처럼)
- cbind 이용 (matrix처럼)

```r
ID <- 1:20
SEX <- sample( c("M","F"), size=20, replace=TRUE)
AGE <- sample( 20:60, size=20 )
AREA <- sample(c("서울","경기","제주","인천","대전","세종"), size=20, replace=T)
df <- data.frame(ID,SEX,AGE,AREA)
str(df)
# $ 사용 
height <- sample( 150:190, size=20, replace=TRUE)
weight <- sample( 40:100, size=20, replace=T)
df$height <- height
str(df)
# [[]] 사용
df[['weight']] <- weight
str(df)
# cbind 사용
weight2 <- sample( 40:100, size=20, replace=T)
df <- cbind(df, weight2)
str(df)
```

## 행추가

```r
ID <- 1:20
SEX <- sample( c("M","F"), size=20, replace=TRUE)
AGE <- sample( 20:60, size=20 )
AREA <- sample(c("서울","경기","제주","인천","대전","세종"), size=20, replace=T)
df <- data.frame(ID,SEX,AGE,AREA)
str(df)

newrow <- data.frame(ID='21',SEX='F',AGE=30,AREA="인천")
df <- rbind(df, newrow)
str(df)
View(df)
```

## sort()와 order() 함수 
- sort() : 정렬된 값을 리턴
- order() : 정렬된 값들의 index 리턴

```r
df$AGE
sort(df$AGE) # 값을 리턴
order(df$AGE) # 인덱스를 리턴
df$AGE[order(df$AGE)] # 리턴된 인덱스를 접근에 사용.
```

```r
# order()로 리턴된 인덱스 벡터를 df의 정렬에 활용
ord = order(df$AGE)
df[ord,]
```
