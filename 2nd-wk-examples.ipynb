## 체중 조절 예제

```r
#입력과 변수 초기화를 위한 명령들
name = readline('name: ')
curr_weight = readline('weight: ')
curr_weight = as.numeric(curr_weight)
target_weight = readline('target weight: ')
target_weight = as.numeric(target_weight)

while(1) {
  cat('1,000 times jump ropes','\n')
  curr_weight <- curr_weight - 1.0
  if (target_weight >= curr_weight)
    break
}
cat(name, "has a success in weight training", curr_weight)
```

## random weight decrement (난수 생성에 의한 몸무게의 감소)
- runif(n) : n개의 uniform distribution 수 생성 

```r
#입력과 변수 초기화를 위한 명령들
name = readline('name: ')
curr_weight = readline('weight: ')
curr_weight = as.numeric(curr_weight)
target_weight = readline('target weight: ')
target_weight = as.numeric(target_weight)

while(1) {
  cat('1,000 times jump ropes','\n')
  curr_weight <- curr_weight - runif(1)
  if (target_weight >= curr_weight)
    break
}
cat(name, "has a success in weight training", curr_weight)
```

## random number generation (난수 생성에 의한 문열림 동작 시뮬레이션)
- runif(n) : n개의 uniform distribution 수 생성 

```r
open = 0
timer = 30
print('버튼')

while(  timer !=0 ) {
  print('가열')
  timer = timer -1 
} 
print("띵")

open = (runif(1) > 0.8)
print(open)
while(open!=1){
  for( timer2 in 1:60) {
      ; #null statement
  }
  print("띵")

  open = (runif(1) > 0.9)
  print(open)
}
```
