## abline() 추세선 그리기
## lines(vec1,vec2) - 선 그리기 
## points(vec1, vec2) - 점 그리기 

```r
cw <- ChickWeight
cw1 <- cw[cw$Chick==1,]

#str(cw1)
plot(cw1$Time, cw1$weight,col='green')

lm0 <- lm(cw1$weight ~ cw1$Time)
lm0
abline(coef(lm0), lwd=2)
abline(c(50,50),lwd=4)
lines(cw1$Time, cw1$weight+10,col='red')
lines(cw1$Time, cw1$weight, col='blue')
points(cw1$Time-2, cw1$weight + 20, col='magenta')
```

## rect()
- 사각형 하나 또는 여러개를 그림, 채우는 색상, 경계선 색상 등도 지정 가능.
- [rect() 참고페이지](https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/rect)


```r
op <- par(bg = "thistle")
plot(c(100, 250), c(300, 450), type = "n", xlab = "", ylab = "",
     main = "2 x 11 rectangles; 'rect(100+i,300+i,  150+i,380+i)'")
i <- 4*(0:10)
## draw rectangles with bottom left (100, 300)+i
## and top right (150, 380)+i
rect(100+i, 300+i, 150+i, 380+i, col = rainbow(11, start = 0.7, end = 0.1))
rect(240-i, 320+i, 250-i, 410+i, col = heat.colors(11), lwd = i/5)

## Background alternating  ( transparent / "bg" ) :
j <- 10*(0:5)
rect(125+j, 360+j,   141+j, 405+j/2, col = c(NA,0),border = "gold", lwd = 2)
rect(125+j, 296+j/2, 141+j, 331+j/5, col = c(NA,"midnightblue"))
mtext("+  2 x 6 rect(*, col = c(NA,0)) and  col = c(NA,\"m..blue\")")

```

```r
op <- par()
## an example showing colouring and shading
plot(c(100, 200), c(300, 450), type= "n", xlab = "", ylab = "")
rect(100, 300, 125, 350) # transparent
rect(100, 400, 125, 450, col = "green", border = "blue") # coloured
rect(115, 375, 150, 425, col = par("bg"), border = "transparent")
rect(150, 300, 175, 350, density = 10, border = "red", angle=-45)
rect(150, 400, 175, 450, density = 30, col = "blue",angle = -30, border = "transparent")

legend(180, 450, legend = 1:4, fill = c(NA, "green", par("fg"), "blue"),
       density = c(NA, NA, 10, 30), angle = c(NA, NA, 30, -30))

par(op)
```

##polygon()



```r
par(bg="white")
n <- 100
x <- c(0,cumsum(rnorm(n)))
y <- c(0,cumsum(rnorm(n)))
xx <- c(0:n, n:0)
yy <- c(x, rev(y))
plot(xx, yy, type="n", xlab="Time", ylab="Distance")
polygon(xx, yy, col="orange")
title("Distance Between Brownian Motions")
```

## [mtext(): 경계 부분에 텍스트 쓰기](https://www.rdocumentation.org/packages/graphics/versions/3.6.2/topics/mtext)
- mtext(text, side = 3, line = 0, outer = FALSE, at = NA,adj = NA, padj = NA, cex = NA, col = NA, font = NA, …)
- text : 텍스트
- side : 경계면 (1=bottom, 2=left, 3=top, 4=right).
- line : 경계 0에서 시작 바깥쪽으로 어느 위치에서 쓸지
- outer : 더바깥 마진이 있다면, 사용 (보통은 F)
- at : user좌표로 각 문자의 위치를 줌


```r
plot(1:10, (-4:5)^2, main = "Parabola Points", xlab = "xlab")
mtext("10 of them")
for(s in 1:4)
    mtext(paste("mtext(..., line= -1, {side, col, font} = ", s,
          ", cex = ", (1+s)/2, ")"), line = -1,
          side = s, col = s, font = s, cex = (1+s)/2)
mtext("mtext(..., line= 3)", line = 3)
mtext("mtext(..., line= -2, adj = 0)", line = -2, adj = 0)
##--- log axis :
plot(1:10, exp(1:10), log = "y", main = "log =\"y\"", xlab = "xlab")
for(s in 1:4) mtext(paste("mtext(...,side=", s ,")"), side = s)
```

```r
## Contour plotting
## This produces a topographic map of one of Auckland's many volcanic "peaks".
x <- 10*1:nrow(volcano) # 10배 확대한 x축 간격
print(x)
y <- 10*1:ncol(volcano)
lev <- pretty(range(volcano), 10)
print(lev) # height

par(bg = "lightcyan") # 배경색 변경

pin <- par("pin") #
cat("pin=",pin) # The current plot dimensions, (width, height), in inches.

xdelta <- diff(range(x))
ydelta <- diff(range(y))
xscale <- pin[1]/xdelta
yscale <- pin[2]/ydelta

scale <- min(xscale, yscale)
xadd <- 0.5*(pin[1]/scale - xdelta)
yadd <- 0.5*(pin[2]/scale - ydelta)
cat("\n xadd =", xadd,"\n yadd=", yadd,"\n")

plot(numeric(0), numeric(0),xlim = range(x)+c(-1,1)*xadd, ylim = range(y)+c(-1,1)*yadd,    type = "n", ann = FALSE)
usr <- par("usr") # user 사용가능 좌표 벡터 c(x1, x2, y1, y2)
rect(usr[1], usr[3], usr[2], usr[4], col=c('purple'),lwd=3)
contour(x, y, volcano, levels = lev, col="yellow", lty="solid", add=TRUE)

box(col='yellow',lwd=3) # plot 주변에 경계 상자 그리기

title("A Topographic Map of Maunga Whau", font= 4)
title(xlab = "Meters North", ylab = "Meters West", font= 3)
mtext("10 Meter Contour Spacing", side=3, line=0.35, outer=FALSE, at = mean(par("usr")[1:2]), cex=0.7, font=3)


```
